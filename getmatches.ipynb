```python
import pandas as pd, requests
```

```python
#get TI matches
ti_matches = requests.get('https://api.opendota.com/api/leagues/14268/matches')
ti_matches = ti_matches.json()
```

```python
#load all ti matches
matches_df = pd.DataFrame(ti_matches)
matches_df.head()

```

```python
#save json file for each match if not yet saved
all_matches = list(matches_df['match_id'])
saved_matches_df = pd.read_csv('working/saved_matches.csv')
saved_matches = list(saved_matches_df['match_id'])
for match in all_matches:
    if match not in saved_matches:
        match_list = []
        match_data = requests.get(f'https://api.opendota.com/api/matches/{match}')
        match_data = match_data.json()
        match_list.append(match_data)
        match_data_df = pd.DataFrame(match_list)
        match_data_json = match_data_df.to_json(f'matches/{match}.json')
        saved_matches.append(match)
        saved_matches_df=pd.DataFrame(saved_matches,columns=['match_id'])
        saved_matches_df.to_csv('working/saved_matches.csv',index=False)
```

```python
#load all matches to json file
all_matches = list(matches_df['match_id'])
all_matches_data = []
loaded_matches_df = pd.read_csv('working/loaded_matches.csv')
loaded_matches = list(loaded_matches_df['match_id'])
loaded_matches_data_df = pd.read_json('all_matches.json')
all_matches_data.append(loaded_matches_data_df)

for match in all_matches:
    if match not in loaded_matches:
        load_match = pd.read_json(f'matches/{match}.json')
        all_matches_data.append(load_match)
        loaded_matches.append(match)
        loaded_matches_df=pd.DataFrame(loaded_matches,columns=['match_id'])
        loaded_matches_df.to_csv('working/loaded_matches.csv',index=False)
        all_matches_data_df = pd.concat(all_matches_data,ignore_index=True)
        all_matches_data_df.to_json('all_matches.json',orient="records")
```
